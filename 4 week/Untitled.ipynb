```python
a1 = int(input())
b1 = int(input())#симкарта
a2 = int(input())
b2 = int(input()) #конверт

#сортируем размеры конвертов и симкарт

if a1 <= b1:
    min1 = a1
    max1 = b1
else:
    min1 = b1
    max1 = a1
    
if a2 <= b2:
    min2 = a2
    max2 = b2
else:
    min2 = b2
    max2 = a2

#проверяем, помещается или нет
if min1<=min2 and max1<=max2:
    print('YES')
else:
    print("NO")
```
