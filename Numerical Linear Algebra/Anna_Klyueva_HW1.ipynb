## Solutions for Problem 1 (Theoretical tasks)

__1.1__ We know that for Euclidean norm $\|x\|_2 = \sqrt{x_1^2 + ...+x_n^2} = \sqrt{x^Tx}$ and for $l_{\infty}$-norm $\|x\|_{\infty} = max_{1\leq i\leq n}|x_i|.$
- It is obvious that $\|x\|_{\infty} \leq \|x\|_2$, as $x_{max} = max_{1\leq i\leq n}|x_i|$ is always less or equal to $\sqrt{x_1^2 +...+ x_{max}^2+...+x_n^2}$.
- $\|x\|_{2}$ takes maximum value if $x_1=...=x_n=x_{max}$. In that case $\|x\|_{\infty} = \sqrt n |x_{max}|$. In other cases $\|x\|_{2}$ will be even less then this value. Hence, if we multiply $\|x\|_{\infty}$ by $\sqrt{n}$   then $\|x\|_{\infty} = \sqrt n |x_{max}|$. 

 $\Rightarrow ~~\frac {1}{\sqrt{n}} \|x\|_2 \leq \|x\|_{\infty} \leq \|x\|_{2} $ 
 
___Answer:___  $ ~C_1 =\frac {1}{\sqrt{n}}, ~ C_2 = 1 $.

__1.2__ From the lecture we know that:
- $(A,B)_F = trace (A^*B), ~~ \|A\|_F = \sqrt{(A,A)_F} = \sqrt{trace(A^*A)}$;
- $U^*U= UU^* = I$, where $U$ is a unitary square matrix;
- Trace is invariant under the cyclic permutation.

Then:
- $ \|UA\|_F = \sqrt{trace((UA)^*UA)} =\sqrt{trace(A^*U^*UA)} = \sqrt{trace(A^*A)} = \|A\|_F $ 
- $ \|AU\|_F = \sqrt{trace((AU)^*AU)} =\sqrt{trace(U^*A^*AU)} = \sqrt{trace(UU^*A^*A)}= \sqrt{trace(A^*A)} = \|A\|_F $

q.e.d.

__1.3__ From the lecture we know that:
- $\|A\|_2 = \sqrt{\lambda_{max}(A^*A)}$;
- Eigenvalues are invariant under transpose operation;
- $U^T$ is also unitary if $U$ is unitary.

Then:
- $\|UA\|_2 = \sqrt{\lambda_{max}((UA)^*UA)} =\sqrt{\lambda_{max}(A^*U^*UA)} = \sqrt{\lambda_{max}(A^*A)} = \|A\|_2 $ 
- $ \|AU\|_2 = \|(AU)^T\|_2 =\|U^TA^T\|_2= \sqrt{\lambda_{max}((U^TA^T)^*U^TA^T)}=\sqrt{\lambda_{max}((A^T)^*(U^T)^*U^TA^T)} =\sqrt{\lambda_{max}((A^T)^*A^T)} = \|A^T\|_2 = \|A\|_2 $

q.e.d.

__2.1__ During the solution I will use facts that:
- $\|A\|_2 =  \sigma_1(A)$
- $A = USV^*$, where $U$ and $V$ are unitary matrices and $S$ is a diagonal matrix containing in its diagonal the singular values of $A$ and $\sigma_1(A) \geq ...\geq \sigma_r(A) $, where $r$ is $rank(A)$

Then $\|A\|_F = \sqrt{trace(A^*A)} = \sqrt{trace((USV^*)^*USV^*)} = \sqrt{trace(VS^*U^*USV^*)} = \sqrt{trace(S^*S)} = \sqrt{\Sigma_{i = 1}^r(\sigma_i(A))^2} \leq \sqrt{rank(A)}\sigma_1(A) = \sqrt{rank(A)}\|A\|_2$

q.e.d.

__2.2__ If we take identity matrix as $A$ with $rank(A) = k = min (n,m)$ then $\sigma_1(A) =...= \sigma_k(A) = 1$ (it is so for any $n$ and $m$) then $\|A\|_F = \sqrt{\Sigma_{i = 1}^k(\sigma_i(A))^2} = \sqrt{k}\sigma_1(A) = \sqrt{rank(A)}\|A\|_2$

q.e.d.

__2.3__ From one of the previous subproblems (using SVD decomposition)  $\|A\|_F = \sqrt{trace(A^*A)} = \sqrt{\sigma_1^2(A)+...+\sigma_r^2(A)} $ where $r = rank(A)$. We also know that $\|A\|_2 =  \sigma_1(A)$. It can be seen that if $rank(A) = 1$ then $\|A\|_F=\|A\|_2$.

q.e.d.

__2.4__ Using SVD decomposition and the fact that $\|UA\|_F = \|A\|_F$ where $U$ is uniary matrix: $\|AB\|_F = \|USV^*B\|_F = \|SV^*B\|_F \leq \|(\sigma_1(A)I)V^*B\|_F = \sigma_1(A)\|V^*B\|_F = \sigma_1(A)\|B\|_F = \|A\|_2 \|B\|_F~~\Rightarrow~~ \|AB\|_F \leq \|A\|_2 \|B\|_F$.

q.e.d.

## Solutions for Problem 2 (Matrix calculus)

__1.1__ $F(U,V) = \frac{1}{2} \| X - UV\|^2_F = \frac{1}{2}\left(X - UV, X - UV \right)\\
dF = \frac{1}{2}\cdot2\left(X - UV,  - dUV\right) = \left(\left(UV - X\right)V^T, dU\right)\\
\frac {dF}{dU} = \left(UV - X\right)V^T$

__1.2__ $F(U,V) = \frac{1}{2} \| X - UV\|^2_F = \frac{1}{2}\left(X - UV, X - UV \right)\\
dF = \frac{1}{2}\cdot2\left(X - UV,  - UdV\right) = \left(U^T\left (UV - X\right), dV\right)\\
\frac {dF}{dV} = U^T(UV - X)$

__1.3__ To compute these gradients we need to:
- transpose matrix one time;
- make one subtraction;
- make two multiplications. 

The highest complexity has multiplication procedure, it is equal to $\mathcal{O}(n^2k)$ if we implement naÃ¯ve or divide and coquer idea.


The dependance should be linear when we fix $n$ and change only $k$ and quadratic, when we fix $k$ and change $n$. (Let's check it and find out what algoritm (Analytical or using JAX) is faster).

````python
from numpy.linalg import norm
import numpy as np
import jax
import jax.numpy as jnp
import time
from tqdm import tqdm
import timeit
import matplotlib.pyplot as plt
````

````python
#the gradient of the function F with respect to U
def dF_dU(X,U,V):
    return (U@V-X)@np.transpose(V)
 
#the gradient of the function F with respect to V
def dF_dV(X,U,V):
    return np.transpose(U)@(U@V-X)

@jax.jit # Just-in-time compilation
def f(X,U,V):
    return 1/2*((jnp.linalg.norm(X-U@V))**2)

#the gradient of the function using JAX
gradf_u = jax.grad(f, argnums=1, has_aux=False)
gradf_v = jax.grad(f, argnums=2, has_aux=False)


a_u_k = []
a_v_k = []
jax_u_k = []
jax_v_k = []

dimensions = [10,100,500,1000,2000,3000,4000,5000,6000, 7000,9000] 
i = 10000
for j in dimensions:  
    
    x = jax.random.normal(jax.random.PRNGKey(0), (i,i))
    u = jax.random.normal(jax.random.PRNGKey(1),(i,j))
    v = jax.random.normal(jax.random.PRNGKey(10),(j,i))
    
    func1_time = %timeit -o -q dF_dU(x,u,v).block_until_ready()
    a_u_k.append(np.mean(func1_time.all_runs))
    
    func1_time = %timeit -o -q dF_dV(x,u,v).block_until_ready()
    a_v_k.append(np.mean(func1_time.all_runs))

    func1_time = %timeit -o -q gradf_u(x,u,v).block_until_ready()
    jax_u_k.append(np.mean(func1_time.all_runs))

    func1_time = %timeit -o -q gradf_v(x,u,v).block_until_ready()
    jax_v_k.append(np.mean(func1_time.all_runs))
````

````python
plt.figure(figsize=(10, 5))
plt.plot(dimensions,a_u_k,label='Analytical')
plt.plot(dimensions,jax_u_k,label='JAX')
plt.title(r'The dependence of running time on the dimension', fontsize=16)
plt.ylabel(r"time", fontsize=16)
plt.xlabel(r"k (n = 10000)", fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(10, 5))
plt.plot(dimensions,a_v_k,label='Analytical')
plt.plot(dimensions,jax_v_k,label='JAX')
plt.title('The dependence of running time on the dimension', fontsize=16)
plt.ylabel(r"time", fontsize=16)
plt.xlabel(r"k (n = 10000)", fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.show()
````

__Conclusion 1:__ On the two plots above we can see the liner dependance as we expected. We can also see that when we calculate $\frac{dF}{dU}$ and $\frac{dF}{dV}$ the timing of analytical approach and timing of the algorithm using JAX are approximately the same.

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch,verbose=True):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0 and verbose:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
````

### Set parameters for training and print intermediate loss values

````python
log_interval = 50
epochs = 7
````

### Third step: run training with the [Adam](https://arxiv.org/pdf/1412.6980.pdf%20%22%20entire%20document) optimization method

If your laptop is not very fast, you will wait some time till training is finished. 

````python
model = Net()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(1, epochs + 1):
    train(model,  train_loader, optimizer, epoch)
    test(model, test_loader)
````

````python
class New_Net(nn.Module):
    def __init__(self, old_net, r):
        super(New_Net, self).__init__()

        self.fc1 = old_net.fc1

        self.u2, self.s2, self.v2  = self.__create_layer(1000, r, 1000, old_net.fc2)
        self.fc2 = nn.Sequential(self.v2, self.s2, self.u2)
        
        self.u3, self.s3, self.v3  = self.__create_layer(1000, r, 1000, old_net.fc3)
        self.fc3 = nn.Sequential(self.v3, self.s3, self.u3)
        
        self.u4, self.s4, self.v4  = self.__create_layer(1000, r, 1000, old_net.fc4)
        self.fc4 = nn.Sequential(self.v4, self.s4, self.u4)
        
        self.u5, self.s5, self.v5  = self.__create_layer(1000, r, 1000, old_net.fc5)
        self.fc5 = nn.Sequential(self.v5, self.s5, self.u5)

        self.fc6 = old_net.fc6        
        self.ReLU = nn.ReLU()

    def __create_layer(self,input_dim, r, output_dim, fc):
        u_w, s_w, v_w = torch.svd(fc.weight.data)
        u_w, s_w, v_w = u_w[:,:r], torch.diag(s_w[:r]), v_w[:,:r].t()

        v = nn.Linear(input_dim, r, bias=False)
        v.weight = torch.nn.Parameter(v_w)

        s = nn.Linear(r, r, bias=False)
        s.weight = torch.nn.Parameter(s_w)

        u = nn.Linear(r,output_dim, bias=True)
        u.weight = torch.nn.Parameter(u_w)
        u.bias = torch.nn.Parameter(fc.bias)

        return u, s, v

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

````python
def test_new(model, test_loader,verbose=True):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    if verbose:
        print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
              test_loss, correct, len(test_loader.dataset),
              100. * correct / len(test_loader.dataset)))
    return 100. * correct / len(test_loader.dataset)
````

````python
import jax.numpy as jnp
import jax
import numpy as np
import random

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)

    def hermione_computes_function(self, x):
        return jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))
        
    def harry_answers(self):
        G = []
        m = 95
        grad_game = jax.jacfwd(self.hermione_computes_function)
        for key in range(0,m):
            x = jax.random.uniform(jax.random.PRNGKey(key), (100,), maxval=100., minval=0., dtype=jnp.float32)
            gradf = grad_game(x)
            G.append(list(gradf))

        G = np.array(G)/np.sqrt(m)
        r = np.linalg.matrix_rank(G)

        (values,counts) = np.unique(r,return_counts=True)

        return values[np.argmax(counts)]
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
