```python
from sklearn import ensemble, model_selection, metrics, datasets, tree, model_selection
import sklearn
import numpy as np
import pandas as pd
```

```python
digits = sklearn.datasets.load_digits()
```

```python
n = digits.data.shape[1]
d = int(n**(1/2))
d
```

```python
import matplotlib.pyplot as plt 
plt.gray() 
plt.matshow(digits.images[0]) 
plt.show()
```

```python
X = digits.data
y = digits.target
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

```python
crossvalscore = sklearn.model_selection.cross_val_score(clf, X, y, cv = 10)
crossvalscore.mean() #ответ на 1 задание
```

```python
b_c = ensemble.BaggingClassifier(n_estimators = 100)
b_c.fit(X,y)
```

```python
crossvalscore = sklearn.model_selection.cross_val_score(b_c, X, y, cv = 10)
crossvalscore.mean() #ответ на 2 задание
```

```python
b_c_1 = ensemble.BaggingClassifier(n_estimators = 100, max_features = 8)
b_c_1.fit(X,y)
```

```python
crossvalscore = sklearn.model_selection.cross_val_score(b_c_1, X, y, cv = 10)
crossvalscore.mean()
```

```python
clf_2 = tree.DecisionTreeClassifier(max_features = d)
clf_2.fit(X, y)
b_c_2 = ensemble.BaggingClassifier(base_estimator = clf_2, n_estimators = 100)
b_c_2.fit(X,y)
```

```python
crossvalscore = sklearn.model_selection.cross_val_score(b_c_2, X, y, cv = 10)
crossvalscore.mean() #ответ на 4 задание
```

```python
n = range(1, 300, 15)
i = 0
score = []
while i < 20:
    RFC = sklearn.ensemble.RandomForestClassifier(n_estimators=n[i])
    RFC.fit(X,y)
    crossvalscore = sklearn.model_selection.cross_val_score(RFC, X, y, cv = 10)
    tmp = crossvalscore.mean()
    score.append(tmp)
    i = i + 1
score
```

```python
n = [5,10,100]
i = 0
score = []
while i < 3:
    RFC = sklearn.ensemble.RandomForestClassifier(max_depth=n[i])
    RFC.fit(X,y)
    crossvalscore = sklearn.model_selection.cross_val_score(RFC, X, y, cv = 10)
    tmp = crossvalscore.mean()
    score.append(tmp)
    i = i + 1
score
```
